```python
from transformers import ClapModel, ClapProcessor

model = ClapModel.from_pretrained("laion/clap-htsat-unfused")
processor = ClapProcessor.from_pretrained("laion/clap-htsat-unfused")
```

```python
import librosa
import torch

# Пример текста и аудио
texts = ["A cat meowing", "Dog barking"]
audio, sr = librosa.load("/content/myaukane-domashney-koshki-37030.wav", sr=48000)  # Важно: 48kHz для CLAP

# Подготовка входных данных
inputs = processor(
    text=texts,
    audios=audio,
    return_tensors="pt",
    sampling_rate=sr,
    padding=True
)

# Получение эмбеддингов
with torch.no_grad():
    outputs = model(**inputs)

# Нормализованные эмбеддинги
text_emb = outputs.text_embeds
audio_emb = outputs.audio_embeds
```

```python
# Вычисление сходства (cosine similarity)
similarity = torch.nn.functional.cosine_similarity(
    audio_emb.unsqueeze(1),
    text_emb.unsqueeze(0),
    dim=-1
)

print("Similarity matrix:")
print(similarity)
```
