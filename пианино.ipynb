```python

from IPython.display import Javascript, display
import base64
def play_sound_auto(file_path):

    with open(file_path, 'rb') as f:
        audio_data = f.read()
    b64_audio = base64.b64encode(audio_data).decode('utf-8')


    js_code = f"""
    var audio = new Audio('data:audio/mp3;base64,{b64_audio}');
    audio.autoplay = true;
    audio.controls = false;
    audio.style.display = 'none';
    document.body.appendChild(audio);
    """
    display(Javascript(js_code))
while True:
  user = str(input())
  if user == 'stop':
    break
  for i in user:
    if i == '1':
      play_sound_auto('/content/re-note-sound.mp3')
    elif i == '2':
      play_sound_auto('/content/c-note-sound.mp3')
    elif i == '3':
      play_sound_auto('/content/fa-note-sound.mp3')
    elif i == '4':
      play_sound_auto('/content/the-sound-of-the-note-do.mp3')
    elif i == '5':
      play_sound_auto('/content/the-sound-of-the-note-la.mp3')
    elif i == '6':
      play_sound_auto('/content/the-sound-of-the-note-mi.mp3')
```
